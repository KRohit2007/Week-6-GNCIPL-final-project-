## 1. Data loading and initial inspection

Load the dataset and perform initial checks (head, info, describe, missing values) on all columns.

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("/content/nhic (1).csv")

# Display the first few rows
display(df.head())

# Check data types and non-null values
df.info()

# Summarize descriptive statistics
display(df.describe(include='all'))

# Check for missing values
print("\nMissing values per column:")
print(df.isnull().sum())
```

## 2. Handle data types

Ensure all columns have appropriate data types, especially converting date columns.

```python
# Convert date columns to datetime objects
df['insurance_apply_date'] = pd.to_datetime(df['insurance_apply_date'])
df['insurance_claimed_date'] = pd.to_datetime(df['insurance_claimed_date'])

# Display the data types to confirm the conversion
df.info()
```

## 3. Analyze distributions

Visualize the distributions of both numerical and categorical variables.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Distribution of numerical variables
df.hist(figsize=(15, 10), bins=20)
plt.suptitle('Histograms of Numerical Variables', y=0.95, fontsize=16)
plt.show()

# Distribution of categorical variables
# Plot 'region' separately as it's a suitable categorical column
plt.figure(figsize=(10, 6))
sns.countplot(x='region', data=df, palette='viridis', hue='region', legend=False)
plt.title('Distribution of Region', fontsize=14)
plt.xlabel('Region')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()


# Distribution of binary categorical variables (sex, smoker, insuranceclaim)
binary_cols = ['sex', 'smoker', 'insuranceclaim']
for col in binary_cols:
    # Ensure binary columns exist in the dataframe and are suitable for plotting
    if col in df.columns:
        # Convert to category type for robust plotting
        df[col] = df[col].astype('category')
        plt.figure(figsize=(6, 4))
        sns.countplot(x=col, data=df, palette='plasma', hue=col, legend=False)
        plt.title(f'Distribution of {col}', fontsize=14)
        plt.xlabel(col)
        plt.ylabel('Count')
        plt.show()
    else:
        print(f"Skipping countplot for column '{col}' due to not found in the dataframe.")
```

## 4. Explore relationships

Investigate relationships between key variables using visualizations, focusing on correlations and potential dependencies. This will include analyzing the relationships between numerical variables through correlation matrices and scatter plots, and exploring how categorical variables relate to numerical ones using techniques like box plots or violin plots.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Correlation matrix for numerical variables
numerical_df = df.select_dtypes(include=np.number)
correlation_matrix = numerical_df.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix of Numerical Variables', fontsize=16)
plt.show()

# Scatter plots for key numerical relationships (example: bill_amount vs age, bill_amount vs bmi)
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.scatterplot(x='age', y='bill_amount', data=df, alpha=0.6)
plt.title('Bill Amount vs. Age')
plt.xlabel('Age')
plt.ylabel('Bill Amount')

plt.subplot(1, 2, 2)
sns.scatterplot(x='bmi', y='bill_amount', data=df, alpha=0.6)
plt.title('Bill Amount vs. BMI')
plt.xlabel('BMI')
plt.ylabel('Bill Amount')

plt.tight_layout()
plt.show()


# Relationships between bill_amount and categorical variables (smoker, region, sex, insuranceclaim)
categorical_vars_to_plot = ['smoker', 'region', 'sex', 'insuranceclaim']

for col in categorical_vars_to_plot:
    if col in df.columns:
        plt.figure(figsize=(8, 5))
        sns.boxplot(x=col, y='bill_amount', data=df, palette='viridis', hue=col, legend=False)
        plt.title(f'Bill Amount vs. {col}', fontsize=14)
        plt.xlabel(col)
        plt.ylabel('Bill Amount')
        plt.show()
    else:
        print(f"Skipping plot for column '{col}' due to not found in the dataframe.")
```

## 5. Identify and handle outliers

Use techniques like z-scores or box plots to identify potential outliers in numerical features, highlighting them in the visualizations.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Select the numerical columns from the DataFrame df.
numerical_cols = df.select_dtypes(include=[np.number]).columns.tolist()
# Exclude patient_id and region_code as they are identifiers/codes
if 'patient_id' in numerical_cols:
    numerical_cols.remove('patient_id')
if 'region_code' in numerical_cols:
    numerical_cols.remove('region_code')


# 2. Calculate the absolute Z-scores for each data point in the selected numerical columns.
z_scores = np.abs((df[numerical_cols] - df[numerical_cols].mean()) / df[numerical_cols].std())

# 3. Identify data points where any of the absolute Z-scores are greater than a threshold (e.g., 3) and consider these as potential outliers.
outliers = df[(z_scores > 3).any(axis=1)]

# 4. Print the number of potential outliers detected.
print(f" Potential anomalies detected: {len(outliers)}")

# 5. Display the head of the DataFrame containing the potential outliers, including relevant columns.
cols_to_display = [col for col in ['patient_id', 'bill_amount', 'region', 'charges'] if col in outliers.columns]
if len(cols_to_display) > 0:
    display(outliers[cols_to_display].head())
else:
    print("No matching columns found to display in outliers subset.")


# 6. Generate a box plot for the 'bill_amount' column to visually inspect for outliers.
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='bill_amount', color='red')
plt.title(' Outlier Check: Bill Amount', fontsize=14)
plt.xlabel('Bill Amount')
plt.show()
print(" Interpretation: Dots outside whiskers are potential outliers in bill amount.")
```
